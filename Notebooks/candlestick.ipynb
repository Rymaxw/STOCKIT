## Import Library

```python
import sys
import os
import warnings
import pandas as pd
warnings.filterwarnings('ignore')

# Tambah root project ke path
ROOT = os.path.abspath('..')
if ROOT not in sys.path:
    sys.path.insert(0, ROOT)

from Utils.candlestick import (
    load_stock_data,
    plot_candlestick,
    plot_candlestick_comparison,
)
```

## Load Data



```python
TICKERS = ['AAPL', 'MSFT', 'NVDA', 'TSLA', 'GOOGL']

dfs = load_stock_data(TICKERS, period='1y', verbose=True)

print()
print('Ticker  | Rows | Kolom')
print('-' * 40)
for t, df in dfs.items():
    print(f'{t:7s} | {len(df):4d} | {list(df.columns)}')

dfs['AAPL'].tail(3)
```

## AAPL: Volume + MA20 + MA50 

```python
fig = plot_candlestick(
    dfs['AAPL'],
    ticker='AAPL',
    title='Apple Inc. — 1 Tahun Terakhir',
)
fig.show()
```

## NVDA: Semua Moving Average

```python
fig = plot_candlestick(
    dfs['NVDA'],
    ticker='NVDA',
    title='NVIDIA Corp. — MA20 + MA50 + MA200',
    show_ma200=True,
)
fig.show()
```

## TSLA: 6 Bulan Terakhir, Tanpa Volume

```python
six_months_ago = (pd.Timestamp.today() - pd.DateOffset(months=6)).strftime('%Y-%m-%d')
print('Filter dari:', six_months_ago)

fig = plot_candlestick(
    dfs['TSLA'],
    ticker='TSLA',
    title='Tesla Inc. — 6 Bulan Terakhir',
    show_volume=False,
    date_from=six_months_ago,
)
fig.show()
```

## MSFT: Candlestick Murni (Tanpa MA)

```python
fig = plot_candlestick(
    dfs['MSFT'],
    ticker='MSFT',
    title='Microsoft Corp. — Candlestick Murni',
    show_ma20=False,
    show_ma50=False,
    show_volume=True,
)
fig.show()
```

## Perbandingan 5 Saham 

```python
fig = plot_candlestick_comparison(
    dfs,
    title='AAPL vs MSFT vs NVDA vs TSLA vs GOOGL',
    height=520,
)
fig.show()
```

## AAPL vs NVDA: 3 Bulan Terakhir

```python
three_months_ago = (pd.Timestamp.today() - pd.DateOffset(months=3)).strftime('%Y-%m-%d')

fig = plot_candlestick_comparison(
    {'AAPL': dfs['AAPL'], 'NVDA': dfs['NVDA']},
    title='AAPL vs NVDA — 3 Bulan Terakhir',
    date_from=three_months_ago,
    height=480,
)
fig.show()
```
