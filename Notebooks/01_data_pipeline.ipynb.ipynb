```python
# import pandas as pd
import yfinance as yf
import pandas as pd
import os 
from IPython.display import display
```

```python
def uji_tarik_data_saham(daftar_ticker, period="10y"):
    data_sukses = {}
    data_gagal = []

    # check folder untuk program
    os.makedirs("../data/processed", exist_ok=True)
    
    for ticker in daftar_ticker:
        try:
            saham = yf.Ticker(ticker)
            df = saham.history(period=period, interval="1d")
            
            if df.empty:
                print(f"Gagal: Data '{ticker}' kosong.")
                data_gagal.append(ticker)
                continue
                
            kolom_wajib = ['Open', 'High', 'Low', 'Close', 'Volume']
            df_base = df[kolom_wajib].copy()
            
            resample_logic = {
                'Open': 'first',
                'High': 'max',
                'Low': 'min',
                'Close': 'last', 
                'Volume': 'sum'
            }
            
            df_weekly = df_base.resample('W').agg(resample_logic).dropna()
            df_monthly = df_base.resample('ME').agg(resample_logic).dropna()
            df_yearly = df_base.resample('YE').agg(resample_logic).dropna()

            # Save to parquet
            df_weekly.to_parquet(f"../data/processed/{ticker}_weekly.parquet")
            df_monthly.to_parquet(f"../data/processed/{ticker}_monthly.parquet")
            df_yearly.to_parquet(f"../data/processed/{ticker}_yearly.parquet")
            
            data_sukses[ticker] = {
                'raw_daily': df_base,
                'weekly': df_weekly,
                'monthly': df_monthly,
                'yearly': df_yearly
            }
            
            print(f"Sukses '{ticker}' -> W: {len(df_weekly)} baris | M: {len(df_monthly)} baris | Y: {len(df_yearly)} baris")
            
        except Exception as e:
            print(f"Terjadi Kesalahan '{ticker}': {e}")
            data_gagal.append(ticker)
            
    return data_sukses, data_gagal
```
